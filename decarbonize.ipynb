Calculating costs

```python
# New column name
new_col_name = 'fuel_costs($/yr)'

# Create the new column by multiplying col1 and col2
vehicles_fuels_df[new_col_name] = vehicles_fuels_df['Yearly range (km)'] * vehicles_fuels_df['Consumption (unit_fuel/km)'] * vehicles_fuels_df['Cost ($/unit_fuel)']

vehicles_fuels_df.head()
```

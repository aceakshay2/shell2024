```python
# Calculate the cumulative sum
cost_profiles_df['cumulative Insurance Cost %'] = cost_profiles_df['Insurance Cost %'].cumsum()
# Calculate the cumulative sum
cost_profiles_df['cumulative Maintenance Cost %'] = cost_profiles_df['Maintenance Cost %'].cumsum()

cost_profiles_df["Total cumulative Insurance Cost % + Maintenance Cost %"] = cost_profiles_df['cumulative Maintenance Cost %'] + cost_profiles_df['cumulative Insurance Cost %'] 


new_order = ["End of Year" , "Resale Value %" , "Insurance Cost %" , "Maintenance Cost %" ,"cumulative Insurance Cost %" , "cumulative Maintenance Cost %" , "Total cumulative Insurance Cost % + Maintenance Cost %" ,"Net cost"]
cost_profiles_df = cost_profiles_df.reindex(columns=new_order)

initial_cost = 100
# Create the new column by multiplying col1 and col2
cost_profiles_df["Net cost"] = (initial_cost * (1-(cost_profiles_df['Resale Value %']/100))) + (initial_cost * cost_profiles_df['Total cumulative Insurance Cost % + Maintenance Cost %']/100) 
cost_profiles_df.head(30)
```

--------------------------------Conclusions-------------------------------- 
1. We are going to write a heuristic that first satisfies the yearly distace demand , chossing the cheapest 1styrcost($) while it satisfies the distance and size constraint.
2. We calculate the total carbon emmissions of the fleet , if it exceeds the yearly amount, We will see what to do :)

```python
import math


distance_mapping = {
      "D1": [f"D{i}" for i in range(1,5)],  # All values from 1 and above
      "D2": [f"D{i}" for i in range(2,5)],  # Values between 1 (exclusive) and 3 (exclusive)
      "D3": [f"D{i}" for i in range(3,5)],  # All values from 3 and above
      "D4": [f"D{i}" for i in range(4,5)],  # No specific range for D4 (assuming all remaining values)
  }
print(distance_mapping)

for year in range(2023 , 2039):
  filtered_demand_df = demand_df[demand_df['Year'] == year]
  filtered_vehicles_fuels_df = vehicles_fuels_df[vehicles_fuels_df['Year'] == year]
  filtered_carbon_emissions = carbon_emissions_df[carbon_emissions_df['Year'] == year]['Carbon emission CO2/kg'].iloc[0] 
  # print(f"carbon emissions {filtered_carbon_emissions_df}")
  solution = []
  carbon_emissions = 0
  for index, row in filtered_demand_df.iterrows():
    size_bucket = row["Size"]
    distance_bucket = row["Distance"]
    demand = row["Demand (km)"]
    print(size_bucket , distance_bucket ,demand  )

    filtered_vehicles_fuels_size_bucket_df = filtered_vehicles_fuels_df[filtered_vehicles_fuels_df['Size'] == size_bucket]
    filtered_vehicles_fuels_size_bucket_df["cost per km ($/km)"] = filtered_vehicles_fuels_size_bucket_df["1styrcost($)"] / filtered_vehicles_fuels_size_bucket_df["Yearly range (km)"] 
    filtered_vehicles_fuels_size_bucket_df = filtered_vehicles_fuels_size_bucket_df[filtered_vehicles_fuels_size_bucket_df['Distance'].isin(distance_mapping[f"{distance_bucket}"])]
    
    lowest_cost_row =  filtered_vehicles_fuels_size_bucket_df.sort_values(by='cost per km ($/km)').iloc[0]

    no_of_vehicles = math.ceil(demand / lowest_cost_row["Yearly range (km)"])
    carbon_emissions+=lowest_cost_row["Yearly range (km)"] * no_of_vehicles 
    print(no_of_vehicles)
    lowest_cost_row["no_of_vehicles"] = no_of_vehicles

    solution.append(lowest_cost_row)

  if(carbon_emissions <= filtered_carbon_emissions ):
    print(f"We got a feasible solution for {year}")
  else:
    print(f"We got a infeasible solution for {year}")
```

```python
# from pulp import LpProblem, LpVariable, lpSum, LpStatus
from pulp import LpProblem, LpVariable, lpSum, LpStatus, LpMinimize

# Write the above code in a function
def optimize_vehicle_selection(vehicles, demands, co2_limit):
  # Create optimization problem
  model = LpProblem(name="Vehicle_Selection", sense=LpMinimize)

  # Decision variables (number of vehicles of each type to buy)
  buy_vehicles = LpVariable.dicts("Buy", vehicles.keys(), lowBound=0, cat="Integer")

  # Objective function (minimize total cost)
  model += lpSum([buy_vehicles[v] * vehicles[v]['1styrcost($)'] for v in vehicles])

  # Demand satisfaction constraint (ensure total demand is met)
  model += lpSum([buy_vehicles[v] * vehicles[v]['Yearly range (km)'] for v in vehicles]) >= demands

  # CO2 emission constraint (total emission should not exceed limit)
  model += lpSum([buy_vehicles[v] * vehicles[v]['Emissions(co2/yr)'] for v in vehicles]) <= co2_limit

  # Solve the problem
  status = model.solve()  # Store solution status

  # Calculate the co2 credit and demand credit
  co2_credit = co2_limit - sum([buy_vehicles[v].value() * vehicles[v]['Emissions(co2/yr)'] for v in vehicles])
  demand_credit = demands - sum([buy_vehicles[v].value() * vehicles[v]['Yearly range (km)'] for v in vehicles])


  # write a code to print the co2 credit and demand credit
  print("CO2 Credit:", co2_limit - sum([buy_vehicles[v].value() * vehicles[v]['Emissions(co2/yr)'] for v in vehicles]))
  print("Demand Credit:", demands - sum([buy_vehicles[v].value() * vehicles[v]['Yearly range (km)'] for v in vehicles]))
  
  # Store the vehicle details
  vehicle_details = {v: int(var.value()) for v, var in buy_vehicles.items() if var.value() > 0}


  # Print solution
  if status == 1:
    print("Minimum Cost:", LpStatus[status], model.objective.value())
    for v, var in buy_vehicles.items():
      if var.value() > 0:
        print(f"Buy {int(var.value())} units of vehicle {v}")
  else:
    print("The problem is infeasible or unbounded")

  # Return the co2 credit, demand credit, and vehicle details
  return {"co2_credit": co2_credit, "demand_credit": demand_credit, "vehicle_details": vehicle_details}

# # Define data (replace with your actual data)
vehicles = {'Diesel_S1_2023': {'1styrcost($)': 20000, 'Yearly range (km)': 1345, 'Emissions(co2/yr)': 1500},
            'Electric_S2_2024': {'1styrcost($)': 30000, 'Yearly range (km)': 1690, 'Emissions(co2/yr)': 50}}
demands = 12000
co2_limit = 10000  # Total CO2 emission limit (units)

# Call the function 
optimize_vehicle_selection(vehicles, demands, co2_limit)
```
